<h3 style="color:orange;">2.0. Planejamento de Solução</h3>
<ul>
    <li>
        <h5 style="color:green;">2.1. Planejamento do Produto Final:</h5>
        <ul>
            <li>
                <span style="color:gray;">O que eu vou entregar? (Planilha, Texto, Email, Modelos de ML, ...)</span>
                <ul>
                    <li><span style="color:gray;">R: Texto com as respostas.</span></li>
                </ul>
            </li>
            <li>
                <span style="color:gray;">Como vai ser a entrega?</span>
                <ul>
                    <li>
                        <span style="color:gray;">R: Perguntas | Respostas.</span>
                        <br>
                        <span style="color:gray;">Por exemplo:</span>
                        <ol>
                            <li><span style="color:gray;">Quantas casas estão disponíveis para compra?</span>
                                <ul><li><span style="color:gray;">R: 2300 imóveis disponíveis para compra.</span></li></ul>
                            </li>
                            <li><span style="color:gray;">Quantos atributos as casas possuem? (nº de quartos, nº de garagens, m², vistas pro mar)</span>
                                <ul><li><span style="color:gray;">R: 10 atributos.</span></li></ul>
                            </li>
                            <li><span style="color:gray;">Quais são os atributos?</span></li>
                            <li><span style="color:gray;">Qual a casa mais cara do portfólio (casa com maior valor)?</span></li>
                            <li><span style="color:gray;">Qual a casa com maior nº de quartos?</span></li>
                        </ol>
                    </li>
                </ul>
            </li>
        </ul>
    </li>
    <li>
        <h5 style="color:green;">2.2. Planejamento do Processo</h5>
        <ul>
            <li><span style="color:gray;">Onde está a informação? (Excel, BD, API, Manual)</span>
                <ul>
                    <li><span style="color:gray;">R: <a href="https://www.kaggle.com/datasets/harlfoxem/housesalesprediction?select=kc_house_data.csv" target="_blank">Link do Artigo</a></span></li>
                </ul>
            </li>
            <li><span style="color:gray;">Como coletar essas informações? (SQL, Python, Streamlit, ...)</span>
                <ul>
                    <li><span style="color:gray;">R: Download arquivo csv.</span></li>
                </ul>
            </li>
            <li>
                <span style="color:gray;">Responder as perguntas?</span>
                <ol>
                    <li><span style="color:gray;">Quantas casas estão disponíveis para compra?</span>
                        <ul><li><span style="color:gray;">R: Contar o nº de linhas no conjunto de dados.</span></li></ul>
                    </li>
                    <li><span style="color:gray;">Quantos atributos as casas possuem? (nº de quartos, nº de garagens, m², vistas pro mar)</span>
                        <ul><li><span style="color:gray;">R: Contar o nº de colunas do conjunto de dados.</span></li></ul>
                    </li>
                    <li><span style="color:gray;">Quais são os atributos?</span>
                        <ul><li><span style="color:gray;">R: Mostrar o nome das colunas (automático).</span></li></ul>
                    </li>
                    <li><span style="color:gray;">Qual a casa mais cara do portfólio (casa com maior valor)?</span>
                        <ul><li><span style="color:gray;">R: Ordenar as linhas pela coluna de preço (atributo).</span></li></ul>
                    </li>
                    <li><span style="color:gray;">Qual a casa com maior nº de quartos?</span>
                        <ul><li><span style="color:gray;">R: Ordenar as linhas pela coluna de nº de quartos (atributo).</span></li></ul>
                    </li>
                </ol>
            </li>
        </ul>
    </li>
    <li>
        <h5 style="color:green;">2.3. Planejamento das Ferramentas</h5>
        <ul>
            <li>
                <span style="color:gray;">Quais ferramentas eu posso usar?</span>
                <ol>
                    <li><span style="color:gray;">Excel</span>
                        <ul>
                            <li><span style="color:gray;">- Fácil de usar.</span></li>
                            <li><span style="color:gray;">- Barato.</span></li>
                            <li><span style="color:gray;">- Muito usado pelos times não técnicos.</span></li>
                            <li><span style="color:gray;">- Poder de processamento limitado.</span></li>
                        </ul>
                    </li>
                    <li><span style="color:gray;">Linguagem de programação</span>
                        <ul>
                            <li><span style="color:gray;">- Desenvolvida para criar software.</span></li>
                            <li><span style="color:gray;">- Processar e Manipular -> Insights.</span></li>
                        </ul>
                    </li>
                </ol>
            </li>
        </ul>
    </li>
</ul>

        <br>
        <ul>

Foi criada por Guido van Rossum e teve sua primeira versão lançada em 1991. <p>Desde então, Python cresceu em popularidade e tornou-se uma das linguagens mais utilizadas em diversas áreas, como desenvolvimento web, ciência de dados, automação, inteligência artificial, aprendizado de máquina, entre outras.</span></li>

</ul>
